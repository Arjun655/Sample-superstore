```python
# distribution the data using the plot
```

```python
fig,axis=plt.subplots(ncols=2,nrows=2,figsize=(10,10))
                      
sns.distplot(data['Sales'],color='red',ax=axis[0][0])
axis[0][0].set_title('Sales Distribution',fontsize=20)
                      
sns.distplot(data['Profit'],color='blue',ax=axis[0][1])
axis[0][1].set_title('Profit Distribution',fontsize=20)
                      
sns.distplot(data['Quantity'],color='green',ax=axis[1][0])
axis[1][0].set_title('Quantity Distribution',fontsize=20)
                      
sns.distplot(data['Discount'],color='black',ax=axis[1][1])
axis[1][1].set_title('Discount Distribution',fontsize=20)

plt.tight_layout()                      
plt.show()
```
